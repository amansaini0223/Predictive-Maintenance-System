```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('ggplot')
```

```python
df = pd.read_csv("../data/raw/ai4i2020.csv")

df.head(5)
```

```python
print("Dataset Shape:")
print(df.shape)
```

```python
df.info()
```

```python
print("Missing Values:")
print(df.isnull().sum())
```

```python
# Statistical Summary
df.describe()
```

```python
df.duplicated().sum()
```

```python
plt.figure(figsize=(6,4))

sns.countplot(
    x='Machine failure',
    data=df
)

plt.title("Machine Failure Distribution")
plt.show()
```

```python
failure_rate = df['Machine failure'].value_counts(normalize=True)*100

print(failure_rate)
```

```python
df.dtypes
```

```python
df["Machine failure"].value_counts()
```

```python
# Correlation Analysis

numeric_df = df.select_dtypes(include=np.number)

plt.figure(figsize=(12,8))

sns.heatmap(
    numeric_df.corr(),
    annot=True,
    cmap='coolwarm'
)

plt.title("Correlation Heatmap")
plt.show()
```

```python
# Numerical Features Distribution

num_cols = [
    'Air temperature [K]',
    'Process temperature [K]',
    'Rotational speed [rpm]',
    'Torque [Nm]',
    'Tool wear [min]'
]

df[num_cols].hist(
    figsize=(15,8),
    bins=20
)

plt.tight_layout()
plt.show()
```

```python
# Boxplot (Outlier Analysis)

for col in num_cols:

    plt.figure(figsize=(6,3))

    sns.boxplot(
        x=df[col]
    )

    plt.title(col)

    plt.show()
```

```python
# Machine Type Distribution

plt.figure(figsize=(6,4))

sns.countplot(
    x='Type',
    data=df
)

plt.title("Machine Type Distribution")
plt.show()
```

```python
# Failure vs Machine Type

plt.figure(figsize=(6,4))

sns.countplot(
    x='Type',
    hue='Machine failure',
    data=df
)

plt.title("Failure by Machine Type")
plt.show()
```

```python
# Pairplot

sample_df = df.sample(1000)

sns.pairplot(
    sample_df,
    hue='Machine failure'
)

plt.show()
```
